data load

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

```python
#data prepeation
y=df['logS']
y
```

```python
x = df.drop('logS', axis=1) # Changed 'log' to 'logS' to match the actual column name
x
```

```python
#split data set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
print(y_lr_train_pred,y_lr_test_pred)
```

```python
#evaluatemodel peformance
x_train
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
print(lr_train_mse)
print(lr_train_r2)
print(lr_test_mse)
print(lr_test_r2)
```

```python
lr_result=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_result.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
```

```python
lr_result
```

evalutemodel performance

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)

y_rf_train_pred=rf.predict(x_train) # Predict on training data to get y_rf_train_pred
y_rf_test_pred=rf.predict(x_test)  # Predict on testing data to get y_rf_test_pred

from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
#rf_test_r2=r2_score(y_test,y_rf_test_pred)  # This line is redundant and can be removed
```

```python
rf_result=pd.DataFrame(['Randam forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_result.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
```

```python
rf_result
```
